```python
# Delete some unuseful columns (Almost all apart from pclass, survived, age, sex)

new_titanic_data = titanic_data.drop(['name','sibsp','parch','ticket','fare',
                                      'cabin','embarked','boat','body','home.dest'], axis=1)
new_titanic_data
```

```python
# Now the dataset is reduced in terms of columns but the data inside are more useful for analysis
# Let's see once again the statistical features of the new/cleaned dataset (This works only for numerical columns)
new_titanic_data.describe()
```

# Conclusion

There's a lot to say at this specific stage. 
1. 1309 people were on the Titanic. More precisely, we have 1309 records. But this doesn't mean that only 1309 people were on board of the Titanic. This analysis is according to the dataset content.

2. Among the persons onboarded, the mean of age is 29.881135, approximately 30. That means that the majority of travelers have an age that converge to 30 years old.

3. The youngest aboard has not even 1 year old. It's a two month baby! And the eldest has 80 years old.

4. From this dataset, it can also be inferred that only 0.381971, approximately 38% of passengers survived.

5. Last but not least, we have 1039 records in pclass and in survived but not in age. The conclusion that can be drawn here is that some records are missing. And this could affect the precision of the whole analysis.
This may be because some passengers did'nt fill their ages during the recording, or that some values have been lost due to a bad data manipulation. Whatever it is, it is not at all good for the analysis.
To deal with this kind of situation, two alternatives can be used : 
- The first one is to use the method fillna() in order to fill the missing values with the mean. Now, this seems to be compromising because in a way or another, the dataset will not be the same. Maybe in the reality, the passenger has less than 30 years old.
- The second one is to use the dropna() method to drop all the missing values. That technique is better than the latter because it doesn't alter the dataset (the real aspect of things) but it reduces its shape, making less data available for analysis. It's always better to loose a few data than compromising the whole dataset by using the first solution. 

```python
# Drop the missing values (only the rows) and store the new dataset into a new variable
new_titanic_data2 = new_titanic_data.dropna(axis=0)
new_titanic_data2
```

```python
new_titanic_data2.shape
```

```python
new_titanic_data2.describe()
```

## Conclusion 
As it can be noticed, the shape of the new dataset has been changed and some of the statistical features also. 
For instance, approximately 41% of the 1046 passengers survived to the accident.
There's no more difference in number of records now and the analysis can still be proceeded.

```python
# Know the number of passengers per class using the method value_counts()

new_titanic_data2['pclass'].value_counts()
```

## Conclusion
According to this result, it can simply be stated that 284 out of 1046 passengers travelled in the first class, while 261 passengers travelled in the second class and finally, the rest of passengers, 501, which is the majority of the passengers travelled in third class.

This last part may imply that most of the passengers belong to a low economic class.

```python
# Group data by some criterias
# Display the mean of all the values grouped by the sex
mean_per_sex = new_titanic_data2.groupby(['sex']).mean()
mean_per_sex
```

## Conclusion 
With this result, it can be stated that approximately women survived more than men.
In fact, approximately 75% of women survived to the Titanic accident whereas only 21% of men survived.

```python
# Group data by sex, pclass 
mean_per_sex_class = new_titanic_data2.groupby(['sex','pclass']).mean()
mean_per_sex_class
```

## Conclusion
The majority of women who survived to the accident travelled in first class, almost the totality.
This percentage gradually decreased for the second and third class where less than 50% of women survived.

Through this, the conclusion that can be drawn is that in the first class, women had more privileges than women and men in every other classes.

At the same time, 35% of men in first class had been saved while only 14% in the second class survived. In the third class, it may seem weird but the number of survivals among the men was greater than the number of survivals in second class.

## Graphical representations of some aspects of the dataset

```python
# Let's start making some useful plots to show how data behave graphically

# Normally, Matplotlib should be imported but Pandas works in collaboration with Matplotlib so this can still work!

# Many types of data graphical representations exist (this can be lines, histograms, bars, charts and many more...)

new_titanic_data2['pclass'].value_counts().plot.line(title="Passengers per Classes")
```

```python
# Same operation but using bar this time
new_titanic_data2['pclass'].value_counts().plot.bar(title = "Passengers per Classes", color="red")
```

```python
new_titanic_data2['age'].value_counts().plot.hist()
```

# Conclusion
Within the two last graphics, data is easily readable and understandable.
